```python
df['visit_date'] = pd.to_datetime(df['visit_date'])                                     # Convert date column to datetime type
```

```python
daily_visitors = df.groupby('visit_date')['visitors'].sum().reset_index()
```

```python

plt.figure(figsize=(15,5))
plt.plot(daily_visitors['visit_date'], daily_visitors['visitors'], color= 'orange')
plt.title("Total Visitors Per Day (All Restaurants)")
plt.xlabel('Date')
plt.ylabel('Number of Visitors')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['day_of_week'] = df['visit_date'].dt.day_name()
```

```python
df.groupby('day_of_week')['visitors'].mean().sort_values(ascending=False)
```

# Day 2: EDA & Visualization

```python
daily_visitors = df.groupby('visit_date')['visitors'].sum().reset_index()

plt.figure(figsize=(15, 5))
plt.plot(daily_visitors['visit_date'], daily_visitors['visitors'])
plt.title("Total Visitors Per Day (All Restaurants)")
plt.xlabel("Date")
plt.ylabel("Number of Visitors")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
dow_avg = df.groupby('day_of_week')['visitors'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.figure(figsize=(10, 5))
sns.barplot(x=dow_avg.index, y=dow_avg.values, palette='viridis')
plt.title("Average Visitors by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Average Visitors")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['visit_date'].dt.month_name()
```

```python
monthly_avg = df.groupby('month')['visitors'].mean()
monthly_avg = monthly_avg.reindex(['January', 'February', 'March', 'April', 'May', 'June','July', 
                                   'August', 'September', 'October', 'November', 'December'])
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values, palette='coolwarm')
plt.title("Average Visitors by Month")
plt.xlabel("Month")
plt.ylabel("Average Visitors")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
one_store_id = df['air_store_id'].unique()[0]
one_store_df = df[df['air_store_id'] == one_store_id]
```

```python
plt.figure(figsize=(15, 5))
plt.plot(one_store_df['visit_date'], one_store_df['visitors'], color='orange')
plt.title(f"Daily Visitors for Store {one_store_id}")
plt.xlabel("Date")
plt.ylabel("Visitors")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['day'] = df['visit_date'].dt.day
pivot = df.pivot_table(index='day_of_week', columns='month', values='visitors', aggfunc='mean')
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title("Heatmap of Avg Visitors by Weekday and Month")
plt.tight_layout()
plt.show()
```
